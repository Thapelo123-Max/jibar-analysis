```python
import pandas as pd
import seaborn as sns
import numpy as np

# Load the data
df = pd.read_csv('jibar_3m.csv')

# Check actual column names
print("Original Columns:", df.columns.tolist())

# Rename correctly
df = df.rename(columns={
    'observation_date': 'Date',
    'IR3TIB01ZAM156N': 'JIBAR_3M'
})

# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Final checks
print("\nShape:", df.shape)
print("\nData types:\n", df.dtypes)
print("\nFirst 5 rows:")
print(df.head())
print("\nSummary Statistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

plt.figure(figsize=(12, 6))

plt.plot(df['Date'], df['JIBAR_3M'], linewidth=2.5, marker='o', markersize=3)

plt.title('3-Month JIBAR Trend (2011 - 2026)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('JIBAR 3M Rate (%)', fontsize=12)
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
